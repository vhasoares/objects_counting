## Cattle Detection using Bag-of-Visual-Words

### Bibliotecas utilizadas nesta implementacao

```python
import cv2  # Biblioteca OpenCV para gerar descritores das imagens
import glob # Biblioteca para trabalhar com diretorios
import time # Biblioteca calcular tempo de treinamento
import numpy as np # Biblioteca de manipulacao de vetores

import matplotlib.image as mpimg # Bibliotecas para plotagem e exibicao de imagens
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.svm import LinearSVC  # Biblioteca para utilizar Classificador
from sklearn.cluster import KMeans # Biblioteca para utilizar K-means
#from scipy.ndimage.measurements import label 


from sklearn.model_selection import train_test_split  # Biblioteca para dividir dataset em trainamento e teste
```

```python
# Variaveis globais

svc = None # variavel para o classificador
detector= None # Variavel para o detector SIFT
kmeans= None # Variavel para o k-means
num_k= 250 # Numero de K
```

## Pré-processamento da imagem

```python
# Funcao que gera um filtro de medias KxK e aplica na imagem G no domínio das frequências
def aplicaFiltroMediaKxK(G, K):
    #Cria o filtro do tamanho da imagem
    filtro = np.zeros(np.shape(G)).astype(np.float32)
    #Preenche o primeiro quadrante KxK com valores de 1/(K*K)
    filtro[0:K,0:K] = 1.0 / (K*K)

    # Retorna a transformada inversa do produto da imagem com o filtro, ambos no dominio de frequencia
    return np.fft.ifft2(np.fft.fft2(G) * np.fft.fft2(filtro)).real



# Funcao de Ajuste de Gamma (aplicado o parametro P)
def gamma(img, P):
    [M, N] = img.shape
    for x in range(M):
        for y in range(N):
            img[x, y] = int(255.0 * ((img[x, y] / 255.0)**(1/P)))
            
    return img


# Função criada para destacar bordas. São analizados os 4 vizinhos do pixel central (vizinhos verticais e horizontais)
# se a diferença entre os pixels estão dentro de um threshold (tsld), o pixel central é interno e recebe valor zero,
# caso contrário, ele é um pixel de borda e recebe um valor proporcional à diferença do pixel central para os vizinhos
def detectaBorda(img, tsld=5):
    [M, N] = img.shape
    img2 = np.zeros((M,N)).astype(np.float32)
    
    for x in range(1, M-1):
        for y in range(1, N-1):
            xc = img[x][y]
            x0 = abs(img[x-1][y] - xc)
            x1 = abs(img[x][y-1] - xc)
            x2 = abs(img[x][y+1] - xc)
            x3 = abs(img[x+1][y] - xc)
            
            if x0 < tsld and x1 < tsld and x2 < tsld and x3 < tsld:
                img2[x,y] = 0
            else:
                img2[x,y] = min(255.0, x0 + x1 + x2 + x3)
                
    return img2



# Funcao para normalizar vetor no intervalo [vmin, vmax] (usado para normalizar de 0 a 255)
def normaliza(vetor, vmin, vmax):
    return vmin + (vmax-vmin) * (vetor - np.min(vetor)) / (np.max(vetor) - np.min(vetor))




# Função que recebe uma imagem inteira e aplica o pipeline de pré-processamento e retorna a imagem resultante pré-processada
def pipeline_preprocessamento(img):

    # Reduz a imagem pra 1/4 do tamanho
    I2 = img[0::4,0::4,:]
    
    # Converte de de BGR (utilizado no openCV) para RGB
    I2 = I2[...,::-1]

    # Filtros para detecção de vegetação (Por enquanto não serão utilizados)
    # Cive
    #I2[:,:,0] = (0.441*I2[:,:,0] - 0.811*I2[:,:,1] + 0.385*I2[:,:,2] + 18.787)
    # ExG
    #I2[:,:,0] = abs(0.1*I2[:,:,1] - I2[:,:,0] -  I2[:,:,2])
    
    
    # Converte imagem para tons de cinza
    I2[:,:,0] = abs(0.21*I2[:,:,0] + 0.72*I2[:,:,1] +  0.07*I2[:,:,2])
    

    # Aplica filtro de média 3x3 para suavizar as transições da vegetação e do terreno,
    # diminuindo a chance das transições serem consideradas animais
    I2 = aplicaFiltroMediaKxK(I2[:,:,0], 3)
    
    # Aplica ajuste de Gamma
    I2 = gamma(I2, 0.5)
    
    # Aplica função de detecção de bordas com threshold 10
    I2 = detectaBorda(I2, tsld=10)

    # Normaliza a imagem para que garanta o intervalo de 0 a 255
    I2 = normaliza(I2, 0.0, 255.0).astype(np.uint8)
    
    # Retorna a imagem pré-processada
    return I2
```

## Funcao para treinamento do classificador

##### Essa função recebe dois conjuntos de imagens, sendo da classe positiva e negativa, e calcula os KeyPoints de cada uma delas com uso do SIFT.

##### Com os Keypoints criados, o algoritmo k-means é aplicado para agrupar esses KeyPoints

##### Depois disso, para cada imagem, os descritores são atribuídos ao cluster mais próximo, onde cada imagem é agora representada pelo histograma de frequência dos grupos em que seus respectivos key-points fazem parte.

##### Esses histogramas são usados para treinar um classificador SVM

```python
def training():
    global detector
    global svc
    global kmeans
    detector=cv2.xfeatures2d.SIFT_create()
    
    # Le o endereço das imagens de "Gado" e "Não-Gado"
    cattle = glob.glob('train_images/Cattle/*.jpg')
    non_cattle = glob.glob('train_images/Not-Cattle/*.jpg')

    
    # Esse trecho é para limitar e utilizar apenas um subconjunto das imagens existentes
    sample_size = 500
    cattle = cattle[0:sample_size]
    non_cattle = non_cattle[0:sample_size]

    print("Loading", len(cattle), "cattle images...")
    print("Loading", len(non_cattle), "non-cattle images...")


    # Cria vetores para armazenas os keypoints que serão encontrados em cada uma das imagens
    catt_features = np.empty((0,128))
    notcatt_features = np.empty((0,128))
    
    # Percorre as imagens "non_cattle" detectando os keypoints e adicionando-os nos vetores criados acima
    for image in non_cattle:
        notcatt=cv2.imread(image,3)
        # Preprocessa a imagem
        notcatt=pipeline_preprocessamento(notcatt)
        
        # Detecta os keypoints
        trainKP,trainDesc = detector.detectAndCompute(notcatt,None)
        if trainDesc is not None:
            notcatt_features = np.vstack((notcatt_features, trainDesc))
            
    # Faz o mesmo agora para as imagens da classe "cattle"
    for image in cattle:
        catt=cv2.imread(image,3)
        # Preprocessa a imagem
        catt=pipeline_preprocessamento(catt)
        trainKP,trainDesc = detector.detectAndCompute(catt,None)
        if trainDesc is not None:
            catt_features = np.vstack((catt_features, trainDesc))

    # Concatena as duas listas de keypoints para agrupar
    X = np.vstack((catt_features, notcatt_features)).astype(np.float64)
    
    # Aplica o algoritmo k-means para encontrar os k centroides de 128 atributos
    kmeans = KMeans(n_clusters=num_k, random_state=0).fit(X)
    
    
    # Agora devemos encontrar os keypoints das imagens novamente e descrever cada imagem no espaço de tamanho num_k
    # que é a quantidade de grupos criados. Ou seja, agora uma imagem é descrita pela quantidade de keypoints de cada
    # grupo que ela possui
    catt_k_train = np.empty((0,num_k))
    notcatt_k_train = np.empty((0,num_k))
    # Faz primeiro para as imagens da classe "non-cattle"
    for image in non_cattle:
        notcatt=cv2.imread(image,3)
        # Preprocessa a imagem
        notcatt=pipeline_preprocessamento(notcatt)
        trainKP,trainDesc = detector.detectAndCompute(notcatt,None)
        # Cria o vetor de tamanho num_k para contabilizar o histograma dos keypoints
        h_img = np.zeros(num_k)
        if trainDesc is not None:
            # usa o kmeans para dizer a qual grupo cada keypoint pertence
            classes = kmeans.predict(trainDesc)
            # Contabiliza quantos keypoints pertencem a cada grupo
            for i in range(num_k):
                h_img[i] = len(np.where(classes == i)[0])
        
            # Normaliza o Vetor pela norma L1
            h_img = h_img / np.linalg.norm(h_img)
            # Adiciona o vetor desta imagem à matriz do conjunto todo
            notcatt_k_train = np.vstack((notcatt_k_train, h_img))
                                                      
    # Faz o mesmo agora para as imagens da classe "cattle"
    for image in cattle:
        catt=cv2.imread(image,3)
        # Preprocessa a imagem
        catt=pipeline_preprocessamento(catt)
        trainKP,trainDesc = detector.detectAndCompute(catt,None)
        h_img = np.zeros(num_k)
        if trainDesc is not None:
            classes = kmeans.predict(trainDesc)
            for i in range(num_k):
                h_img[i] = len(np.where(classes == i)[0])
            h_img = h_img / np.linalg.norm(h_img) # Normaliza o vetor usando norma L1
            catt_k_train = np.vstack((catt_k_train, h_img))
    
    
    # Concatena os vetores das classes "non-cattle" com os vetores das classes "cattle"                                                  
    X = np.vstack((catt_k_train, notcatt_k_train)).astype(np.float64)
    

    # Define o vetor de rótulos do conjunto
    y = np.hstack((np.ones(len(catt_k_train)), np.zeros(len(notcatt_k_train))))

    # Corta o conjunto sub-conjuntos aleatórios de treinamento e testes. O conjunto de testes foi definido como 20%
    rand_state = np.random.randint(0, 100)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=rand_state)


    # Usa classificador linear SVC
    svc = LinearSVC()
    
    # Computa o tempo de treinamento para o SVC
    t=time.time()
    svc.fit(X_train, y_train)
    t2 = time.time()
    print(round(t2-t, 2), 'Segundos para treinar o SVC...')

    # Confere a acurácia do SVC
    print('Acuracia de teste do SVC = ', round(svc.score(X_test, y_test), 4))
```

```python
# Chama a funcao de treinamento do classificador
training()
```

### Função para classificar uma imagem como Cattle ou not-Cattle

```python
# Esta função recebe como parâmetro uma imagem
def predizer(img):
    # Preprocessa a imagem
    img=pipeline_preprocessamento(img)
    # Encontra os keypoints da imagem utilizando o método SIFT
    trainKP,trainDesc = detector.detectAndCompute(img,None)
    h_img = np.zeros((1, num_k))
    # Se há keypoints na imagem, verifica a qual grupo cada keypoint pertence
    if trainDesc is not None:
        classes = kmeans.predict(trainDesc)
        # Conta quantos keypoints cada grupo essa imagem possui
        for i in range(num_k):
            h_img[0,i] = len(np.where(classes == i)[0])
    
        h_img = h_img / np.linalg.norm(h_img) # Normaliza o vetor usando norma L1
        
    # Classifica a imagem como "Cattle" ou "not-Cattle"    
    test_prediction = svc.predict(h_img)
    return test_prediction # 0- Cattle , 1- not-Cattle
```

### Janela deslizante de tamanho fixo para encontrar objetos em uma imagem inteira

```python
# Função recebe uma imagem inteira e a quantidade de pixels que define o tamanho da janela
def find_cattle(img, pix_per_cell):

    # cria uma copia da imagem para desenhar as caixas
    draw_img = np.copy(img)
    draw_img = np.array(draw_img[0::4,0::4,:])
    # Preprocessa a imagem
    img=pipeline_preprocessamento(img)
    # Define a quantidade de janelas terao na imagem
    nxblocks = (img.shape[1] // pix_per_cell)-1
    nyblocks = (img.shape[0] // pix_per_cell)-1 

    # Percorre as janelas variando o eixo X e Y
    for xb in range(nxblocks):
        for yb in range(nyblocks):
            # Computa onde comeca a janela atual
            xleft = xb*pix_per_cell
            ytop = yb*pix_per_cell
               
            # Detecta os keypoints da sub-imagem contida na janela atual    
            trainKP,trainDesc = detector.detectAndCompute(img[ytop:ytop+pix_per_cell-1, xleft:xleft+pix_per_cell-1],None)
            h_img = np.zeros((1, num_k))
            # Calcula o histograma dos keypoints baseados nos grupos do k-means e classifica a sub-imagem
            if trainDesc is not None:
                classes = kmeans.predict(trainDesc)
                for i in range(num_k):
                    h_img[0,i] = len(np.where(classes == i)[0])
            test_prediction = svc.predict(h_img)
            
            # Se a sub-imagem é um gado, desenha um box nela
            if test_prediction == 0:
                xbox_left = np.int(xleft)
                ytop_draw = np.int(ytop)
                cv2.rectangle(draw_img,(xbox_left, ytop_draw),(xbox_left+pix_per_cell,ytop_draw+pix_per_cell),(255,0,0),4) 
                
    return draw_img
```

## Exemplo de classificação de um animal

```python
# Nesta célula, uma imagem de um gado é aberta e a função de predizer é chamada para classificar a imagem
# Um grafico é exibido a seguir com a imagem e o rótulo previsto logo acima.
img = cv2.imread("train_images/Cattle/yes (1208).jpg",3)
img2 = cv2.imread("train_images/Not-Cattle/not (151).jpg",3)

# Prediz a primeira imagem
if predizer(img) == 0:
    title1 = "Cattle"
else:
    title1 = "Not-Cattle"
    
# Prediz a segunda imagem
if predizer(img2) == 0:
    title2 = "Cattle"
else:
    title2 = "Not-Cattle"
    

fig = plt.figure()
fig.set_size_inches(12, 4)
plt.subplot(121)
plt.imshow(img[...,::-1].astype(np.uint8))
plt.title(title1, fontsize=20)


plt.subplot(122)
plt.imshow(img2[...,::-1].astype(np.uint8))
plt.title(title2, fontsize=20)

fig.tight_layout()
```

## Exemplo de detecção de objetos

```python
# Nesta célula, uma imagem é aberta e a função de detecção é chamada
# Ao fim, a imagem com o gado detectado é exibido

img =cv2.imread("DJI_0469.JPG",3)
draw_image = find_cattle(img, 25)
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
plt.subplot(121)
plt.imshow(img[:,:,::-1].astype(np.uint8))
plt.title('Original', fontsize=20)
plt.subplot(122)

plt.imshow(draw_image[:,:,::-1])
plt.title('Com Boxes', fontsize=20)
fig.tight_layout()
```
