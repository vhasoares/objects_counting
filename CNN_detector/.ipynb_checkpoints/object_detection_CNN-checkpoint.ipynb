## Preparação do Modelo

### Variáveis

Localização do grafo com os pesos "congelados", para a tarefa de classificação

Abaixo é o arquivo com os Labels para as duas classes existentes (Cattle e Not-Cattle)

```python
# Modelo treinado
MODEL_NAME = 'cnn_graph'
PATH_TO_CKPT = MODEL_NAME + '/frozen_inference_graph.pb'

# Arquivo com as Labels
PATH_TO_LABELS = os.path.join('cnn_graph', 'object-detection.pbtxt')
NUM_CLASSES = 2
```

### Carrega o Grafo modelo do TensorFlow na memória

```python
detection_graph = tf.Graph()
with detection_graph.as_default():
  od_graph_def = tf.GraphDef()
  with tf.gfile.GFile(PATH_TO_CKPT, 'rb') as fid:
    serialized_graph = fid.read()
    od_graph_def.ParseFromString(serialized_graph)
    tf.import_graph_def(od_graph_def, name='')
```

### Carrega o mapa de rótulos com seus respectivos índices

```python
label_map = label_map_util.load_labelmap(PATH_TO_LABELS)
categories = label_map_util.convert_label_map_to_categories(label_map, max_num_classes=NUM_CLASSES, use_display_name=True)
category_index = label_map_util.create_category_index(categories)
```

### Função para abrir imagem como Numpy

```python
def load_image_into_numpy_array(image):
  (im_width, im_height) = image.size
  return np.array(image.getdata()).reshape(
      (im_height, im_width, 3)).astype(np.uint8)
```

## Detecção a partir de uma lista de imagens

```python
# Lista de imagens a serem detectadas
PATH_TO_TEST_IMAGES_DIR = 'images/input'
TEST_IMAGE_PATHS = [ os.path.join(PATH_TO_TEST_IMAGES_DIR, 'image{}.JPG'.format(i)) for i in range(1, 5) ]
```

```python
# Tamanho da saída das imagens
IMAGE_SIZE = (10, 30)
plt.figure(figsize=IMAGE_SIZE, dpi=75)
# Gerar numero de subplots baseado na quantidade de imagens carregadas (ex.: 4 images gera subplot 410 (4 linhas e 1 coluna))
plotnum = len(TEST_IMAGE_PATHS) * 100 + 10

# Inicia a deteccao de objetos para cada uma das imagens
with detection_graph.as_default():
  with tf.Session(graph=detection_graph) as sess:
    # Faz a predição para cada imagem do conjunto de teste
    for image_path in TEST_IMAGE_PATHS:
      image = Image.open(image_path)
      # Converte a imagem para uma matriz numpy
      image_np = load_image_into_numpy_array(image)
      # Expande as dimenssões pois o modelo espera que as imagens sejam: [1, H, W, 3]
      image_np_expanded = np.expand_dims(image_np, axis=0)
      image_tensor = detection_graph.get_tensor_by_name('image_tensor:0')
      # Cada boxes representa uma parte da imagem onde um objeto foi detectado
      boxes = detection_graph.get_tensor_by_name('detection_boxes:0')
      # Cada Score representa o nível de confiança para cada objeto detectado
      # O Score será mostrado no resultado final na caixa que delimita o objeto
      scores = detection_graph.get_tensor_by_name('detection_scores:0')
      classes = detection_graph.get_tensor_by_name('detection_classes:0')
      num_detections = detection_graph.get_tensor_by_name('num_detections:0')
      # Executa a Deteccao
      (boxes, scores, classes, num_detections) = sess.run(
          [boxes, scores, classes, num_detections],
          feed_dict={image_tensor: image_np_expanded})
        
      #Ignorar as boxes que não são gado
      m, n= np.where(classes == 1)
      num_detections = np.size(n)
      classes = classes[m,n]
      boxes = boxes[m,n]
      scores = scores[m,n]
      boxes = boxes[[0, 1]]
      classes = classes[[0, 1]]
      scores = scores[[0, 1]]
      # Visualizacao do resultado da deteccao
      vis_util.visualize_boxes_and_labels_on_image_array(
          image_np,
          np.squeeze(boxes),
          np.squeeze(classes).astype(np.int32),
          np.squeeze(scores),
          category_index,
          max_boxes_to_draw=1000,
          min_score_thresh=0.1, # Threshold de certeza para considerar objeto como gado ou nao
          use_normalized_coordinates=True,
          line_thickness=4)
      plotnum += 1 # exibe a imagem no proximo subplot
      plt.subplot(plotnum)
    # Mostra o resultado na tela
      plt.imshow(image)#_np)
    # Salva a imagem no arquivo
      cont += 1
plt.savefig('output_images/result.jpg', dpi=75)
      
```
