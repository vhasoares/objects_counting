```python
# Funcao que gera um filtro de medias KxK e aplica na imagem G no domínio das frequências
def aplicaFiltroMediaKxK(G, K):
    #Cria o filtro do tamanho da imagem
    filtro = np.zeros(np.shape(G)).astype(np.float32)
    #Preenche o primeiro quadrante KxK com valores de 1/(K*K)
    filtro[0:K,0:K] = 1.0 / (K*K)

    # Retorna a transformada inversa do produto da imagem com o filtro, ambos no dominio de frequencia
    return np.fft.ifft2(np.fft.fft2(G) * np.fft.fft2(filtro)).real
```

```python
# Funcao de Ajuste de Gamma (aplicado o parametro P)
def gamma(img, P):
    [M, N] = img.shape
    for x in range(M):
        for y in range(N):
            img[x, y] = int(255.0 * ((img[x, y] / 255.0)**(1/P)))
            
    return img
```

```python
# Função criada para destacar bordas. São analizados os 4 vizinhos do pixel central (vizinhos verticais e horizontais)
# se a diferença entre os pixels estão dentro de um threshold (tsld), o pixel central é interno e recebe valor zero,
# caso contrário, ele é um pixel de borda e recebe um valor proporcional à diferença do pixel central para os vizinhos
def detectaBorda(img, tsld=5):
    [M, N] = img.shape
    img2 = np.zeros((M,N)).astype(np.float32)
    
    for x in range(1, M-1):
        for y in range(1, N-1):
            xc = img[x][y]
            x0 = abs(img[x-1][y] - xc)
            x1 = abs(img[x][y-1] - xc)
            x2 = abs(img[x][y+1] - xc)
            x3 = abs(img[x+1][y] - xc)
            
            if x0 < tsld and x1 < tsld and x2 < tsld and x3 < tsld:
                img2[x,y] = 0
            else:
                img2[x,y] = min(255.0, x0 + x1 + x2 + x3)
                
    return img2
            
```

```python
# Funcao para normalizar vetor no intervalo [vmin, vmax] (usado para normalizar de 0 a 255)
def normaliza(vetor, vmin, vmax):
    return vmin + (vmax-vmin) * (vetor - np.min(vetor)) / (np.max(vetor) - np.min(vetor))
```

```python
# Abre uma imagem a ser pré-processada
I = imageio.imread("images/DJI_0030.JPG").astype(np.float32) # Nome da imagem a ser filtrada

# Reduz a imagem pra 1/4 do tamanho
I2 = I[0::4,0::4,:]

# Filtros para detecção de vegetação
# Cive
#I2[:,:,0] = (0.441*I2[:,:,0] - 0.811*I2[:,:,1] + 0.385*I2[:,:,2] + 18.787)
# ExG
#I2[:,:,0] = abs(0.1*I2[:,:,1] - I2[:,:,0] -  I2[:,:,2])

# Aplica filtro de média 3x3 para suavizar as transições da vegetação e do terreno, diminuindo a chance das transições serem
# consideradas animais
I2 = aplicaFiltroMediaKxK(I2[:,:,0], 3)
# Aplica ajuste de Gamma
I2 = gamma(I2, 0.5)
# Aplica função de detecção de bordas com threshold 10
I2 = detectaBorda(I2, tsld=10)

# Normaliza a imagem para que garanta o intervalo de 0 a 255
I2 = normaliza(I2, 0.0, 255.0)
```
